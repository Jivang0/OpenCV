```python
import cv2 as cv
import matplotlib.pyplot as plt 

# load image 
image = cv.imread(r"image\wwwww (1).jpg")
images = cv.cvtColor(image,cv.COLOR_BGR2RGB)


blur3 = cv.GaussianBlur(images,(3,3),0)
blur5 = cv.GaussianBlur(images,(5,5),0)
blur7 = cv.GaussianBlur(images,(7,7),0)

# plot theorginal and blurred images 
titles =["Original","3*3 kernel","5*5 kernel","7*7 kernel"]
images =[image,blur3,blur5,blur7]


plt.figure(figsize=(10,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis("off")
plt.show()
```

```python
# Load image (use a noisy image)
image = cv.imread(r"image\wwwww (1).jpg")
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Apply Median Filter with 5x5 kernel
median_filtered = cv.medianBlur(image, 5)  # 5 = kernel size (must be odd)

# Show original vs median filtered image
titles = ["Original", "Median Filter (5x5)"]
images = [image, median_filtered]

plt.figure(figsize=(8, 6))
for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis("off")
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load and prepare the image
image = cv.imread(r"image\back.webp")  # Replace with your image path
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Function to add salt-and-pepper noise
def add_salt_pepper_noise(img, prob):
    noisy = img.copy()
    black = 0
    white = 255
    threshold = 1 - prob
    rnd = np.random.rand(*img.shape[:2])
    noisy[rnd < prob] = black
    noisy[rnd > threshold] = white
    return noisy

# Add noise
noisy_image = add_salt_pepper_noise(image, 0.05)  # 5% noise

# Apply filters
gaussian_filtered = cv.GaussianBlur(noisy_image, (5, 5), 0)
median_filtered = cv.medianBlur(noisy_image, 5)
bilateral_filtered = cv.bilateralFilter(noisy_image, d=9, sigmaColor=75, sigmaSpace=75)

# Display results
titles = ["Original", "Noisy Image", "Gaussian Blur", "Median Filter", "Bilateral Filter"]
images = [image, noisy_image, gaussian_filtered, median_filtered, bilateral_filtered]

plt.figure(figsize=(12, 10))
for i in range(5):
    plt.subplot(2, 3, i + 1)
    plt.title(titles[i])
    plt.imshow(images[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
